```python
import geobr
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
#Lista dataset
geobr.list_geobr()
```

```python
#vamos selecionar um estado
br=geobr.read_state(year=2019)
br.head()
```

```python
pa=br[br.abbrev_state=='PA']
pa
```

```python
#Plotar os estados Brasileiros e realçar um deles
fig, ax=plt.subplots(figsize=(10,10), dpi=100)

#Definindo as Bases
base= br.plot(color='white', edgecolor='black', ax=ax )
pa.plot(ax=base, color='green')

#Inserindo Titulo
ax.set_title('Brasil', fontsize=20)
ax.axis('on')
```

```python
# Arranjos populacionais

bra = geobr.read_pop_arrangements (year = 2015)
bra.head()
```

```python
para = bra[bra.name_state=='PARÁ']
para
```

```python
#Plotar os estados Brasileiros e realçar um deles
fig, ax=plt.subplots(figsize=(10,10), dpi=100)

#Definindo as Bases
base= bra.plot(color='white', edgecolor='black', ax=ax )
para.plot(ax=base, color='green')

#Inserindo Titulo
ax.set_title('Pará', fontsize=20)
ax.axis('on')
```
