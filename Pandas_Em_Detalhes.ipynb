```python
#Selecionar da linha 1 até o passageiro 10 e quero essas 3 colunas
df.loc[1:10, ['Name','Sex','Age']]
```

#<h3> Trabalhando com o método query()</h3>

```python
#Pesquisar pessoas com idade maior que 20 anos
df.query('Age > 20').head()
```

```python
#Pesquisar pessoas com idade maior que 20 anos do sexo masculino
df.query('Age > 20 & Sex=="male"').head(10)
```

*Usando o operador OR*

```python
# |or (ou)
df.query('Age > 20 | Sex=="male"').head(10)
```

<h4> Operador In e o parÂmetro inplace</h4>

```python
#Valor estar In uma lista de valores
#Eu quero pessoas que embarcaram em C e Q
#Quero alterar o DataFrame pra refletir esse resultado
df.query('Embarked in["C","Q"]', inplace=True)
```
